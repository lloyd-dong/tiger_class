```python
import random
total_number=10000000
in_circle =0
#生成随机数
for i in range(total_number):
    x=random.random()
    y=random.random()
    r2= x**2 + y**2
    #print x, y, r2
    
    #计算圆内点
    if 1 > r2:
        in_circle = in_circle+1
print "圆内点 :",in_circle
# 计算pi
pi = 4.0 * in_circle / total_number 
print "pi=",pi ," 误差=",pi-3.1415926575
```
